# Exercice 1 : Computing Eigen Vector

```python
#Declaring Variables
A = np.array([[0.3, 0.6, 0.1],
              [0.5, 0.2, 0.3],
              [0.4, 0.1, 0.5]])
v = np.array([1/3, 1/3, 1/3])
y = []

#Calculate the eigen vector for 'A'
for i in range(25):
    v_prime = v.dot(A)
    y.append(np.linalg.norm(v_prime - v))
    v = v_prime

#Plot the Euclidien distance between v' and v
plt.plot(range(25), y, c='blue')
plt.show()
```

# Exercice 5 : Test Matrix Symmetricity

```python
#Declaring Test Variables
I3 = np.eye(3)
A = np.array([[0.3, 0.6, 0.1],
              [0.5, 0.2, 0.3],
              [0.4, 0.1, 0.5]])

#Defining the automatoc function
def is_symmetric_automatic(matrix):
    return (matrix == matrix.T).all()
```

```python
#Testing the Function
is_symmetric_automatic(I3)
```

```python
#Testing the Function
is_symmetric_automatic(A)
```

```python
#Defining the Manual Function
def is_symmetric_manual(matrix):
    matrix_T = np.array( [ [0] * len(matrix) ] * len(matrix[0]) )
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            matrix_T[i,j] = matrix[j, i]
    return (matrix == matrix_T).all()
```

```python
#Testing the Function
is_symmetric_manual(I3)
```

```python
#Testing the Function
is_symmetric_manual(A)
```

```python
#Defining the angle variable for the circles
theta = np.linspace(-np.pi, np.pi, 200)

#Plotting the unifromly randomized radius varible for both circles 
for t in theta:
    plt.scatter(10*np.sin(t) + 2*np.random.random(1), 10*np.cos(t) + 2*np.random.random(1), c= 'blue')
    plt.scatter(20*np.sin(t) + 2*np.random.random(1), 20*np.cos(t) + 2*np.random.random(1), c= 'red')

#Showing the Concentric Cirlces Dataset
plt.axis('equal')
plt.show()
```
